## Playing with Neural Networks

```python
import matplotlib.pyplot as plt
import numpy as np 
import keras as kr
```

```python
## Formatting plots
plt.rcParams['figure.figsize'] = (10, 10)
```

### Generating some functions

```python
x = np.linspace(-10.0, 10.0, 2000)
```

```python
y1 = 3*x+5
```

```python
y2= x*x -41
```

```python
y3 = 10*np.sin(2*x)
```

```python
plt.plot(x, y1, label ="Function1")
plt.plot(x, y2, label ="Function2")
plt.plot(x, y3, label ="Function3");
plt.legend();
```

```python
#y = y1 #Linear
y = y2 #Polynomial
#y = y3 #Sine Function

# Create a new neural network.
#Sequential means I will build the layers one by one
m = kr.models.Sequential()

# Add neurons.
#m.add(kr.layers.Dense(1, input_dim=1, activation="linear"))

# Add neurons
m.add(kr.layers.Dense(20, input_dim=1, activation="tanh")) 
m.add(kr.layers.Dense(20, activation="linear"))
m.add(kr.layers.Dense(1, activation='linear')) #one layer because we only want one output
 
# Compile the model.
#Mean squared error = Average squared distance from my prediction to actual value (Standard Deviation calc)
m.compile(loss="mean_squared_error", optimizer="adam")

# Train the model.
# Use this dataset (x, y) fifteen times, and take ten data points each time. 
m.fit(x, y, epochs=20, batch_size=10)

# Run each x value through the neural network.
p = m.predict(x)

# Plot the values.
plt.plot(x, p, 'r-', label='Prediction')
plt.plot(x, y, 'k:', label='Original')
plt.legend();
```
